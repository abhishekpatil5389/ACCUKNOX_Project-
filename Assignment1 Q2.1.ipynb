```python
import subprocess

# Define thresholds
CPU_THRESHOLD = 80
MEMORY_THRESHOLD = 80
DISK_THRESHOLD = 80
PROCESS_THRESHOLD = 100

# Function to check CPU usage
def check_cpu_usage():
    output = subprocess.check_output(['top', '-bn1']).decode('utf-8')
    cpu_line = next(line for line in output.split('\n') if line.startswith('%Cpu'))
    cpu_usage = float(cpu_line.split()[1])
    if cpu_usage > CPU_THRESHOLD:
        print(f"CPU usage is {cpu_usage}% which exceeds the threshold of {CPU_THRESHOLD}%")

# Function to check memory usage
def check_memory_usage():
    output = subprocess.check_output(['free']).decode('utf-8')
    memory_line = next(line for line in output.split('\n') if line.startswith('Mem:'))
    memory_total, memory_used = map(int, memory_line.split()[1:3])
    memory_usage = (memory_used / memory_total) * 100
    if memory_usage > MEMORY_THRESHOLD:
        print(f"Memory usage is {memory_usage}% which exceeds the threshold of {MEMORY_THRESHOLD}%")

# Function to check disk usage
def check_disk_usage():
    output = subprocess.check_output(['df', '-h', '/']).decode('utf-8')
    disk_usage_line = output.split('\n')[1]
    disk_usage = int(disk_usage_line.split()[4].strip('%'))
    if disk_usage > DISK_THRESHOLD:
        print(f"Disk usage is {disk_usage}% which exceeds the threshold of {DISK_THRESHOLD}%")

# Function to check number of running processes
def check_running_processes():
    output = subprocess.check_output(['ps', 'aux']).decode('utf-8')
    num_processes = len(output.split('\n')) - 1
    if num_processes > PROCESS_THRESHOLD:
        print(f"Number of running processes is {num_processes} which exceeds the threshold of {PROCESS_THRESHOLD}")

# Main function
def main():
    # Check CPU usage
    check_cpu_usage()

    # Check memory usage
    check_memory_usage()

    # Check disk usage
    check_disk_usage()

    # Check running processes
    check_running_processes()
```
